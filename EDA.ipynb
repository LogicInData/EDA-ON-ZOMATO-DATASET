# MERGING TWO DATASETS

```python
merged_z = pd.merge(z,c,on = 'Country Code',how = 'left')
```

```python
merged_z
```

# CHECK DATATYPES

```python
merged_z.dtypes
```

```python
merged_z.columns
```

```python
max_base = merged_z['Country'].value_counts()
country_names = merged_z['Country'].value_counts().index
country_values = merged_z['Country'].value_counts().values
```

# PIE CHART - TOP 3 ZOMATO USERS

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct = '%1.2f%%')
```

### **Observation :** Zomato's maximum records or transactions are from India.Right after india USA leads followe by UK.

```python
ratings = merged_z.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

# BARPLOT

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data = ratings)
```

# BARPLOT USING RATING COLOR

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue = 'Rating color',data = ratings)
```

# MAPPING COLORS BASED ON THE RATINGS DATA

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue = 'Rating color',data = ratings,palette =['white','red','orange','green','green'])
```

# OBSERVATION -
1. Not Rated Count - Very High
2. Maximum Rating - Between 2.5 - 3.4

# COUNT PLOT

```python
sns.countplot(x='Rating color',hue = 'Rating color',data = ratings,palette =['white','red','orange','green','green'])
```
